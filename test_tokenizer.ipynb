```python
import json

with open("./assets/count.jsonl", "r") as f:
    count = json.load(f)
```

```python
all_count = sum([c for _, c in count])
tmp_count = 0

remove_pieces = set()

for ids, c in count:
    tmp_count += c
    remove_pieces.add(new_llama_add_chinese_other_model.pieces[ids].piece)
    print((new_llama_add_chinese_other_model.pieces[ids].piece, tmp_count / all_count))

    if tmp_count > all_count * 0.9:
        break
```
